```python
import pandas as pd
```

```python
file_directory='./data_after/'
```

```python
match_data=pd.read_csv("{0}match.csv".format(file_directory))
champion_data=pd.read_csv("{0}champion.csv".format(file_directory))
player_data=pd.read_csv("{0}player.csv".format(file_directory))
```

```python
match_data_index=set(match_data.index)
match_data.tail()
```

```python
lane=['Top','Jungle','Middle','ADC','Support']
player_attr=['W%','KDA','KP','GD10','XPD10','CSD10','DPM']
champ_attr=['P%','B%','W%']
```

```python
columns=['side','result']
side=['side1','side2']
for elem1 in side:
    for elem2 in lane:
        for elem3 in player_attr:
            columns.append('{0} {1} player {2}'.format(elem1,elem2,elem3))
    for elem2 in lane:
        for elem3 in champ_attr:
            columns.append('{0} {1} champ {2}'.format(elem1,elem2,elem3))
        
```

```python
data_body=[]
```

```python
for elem in match_data_index:
    
    
    lst1=[]
    lst2=[]
    after_data1=[]
    after_data2=[]
    temp_data=match_data.loc[elem][match_data.columns.difference(['league'])]
    datas=[temp_data[0:5],temp_data[5:]]
    
    j=0
    for data in datas:
        
            
        result=data.iloc[0]['result']
        side=data.iloc[0]['side']
        team=data.iloc[0]['team']
        year=data.iloc[0]['year']-2000
        season=data.iloc[0]['split']
        players=data['player']
        champions=data['champion']
        position=data['position']
        
        
        if j==0:
            after_data1=[result,side]
        else:
            after_data2=[result,side]
            
            
        cond1=(player_data['Year']==year)
        cond2=(player_data['Season']==season)
        cond5=(player_data['Team']==team)
        
        
        i=0
        
        for player in players:
            cond3=(player_data['Player']==player)
            cond4=(player_data['Pos']==position.iloc[i])
            if j==0:
                lst1.extend(player_data[cond1&cond2&cond3&cond4&cond5][player_attr].iloc[0])   
            else:
                lst2.extend(player_data[cond1&cond2&cond3&cond4&cond5][player_attr].iloc[0])   
            i+=1

        
        i=0
        cond1=(champion_data['Year']==year)
        cond2=(champion_data['Season']==season)
        for champion in champions:
            cond3=(champion_data['Champion']==champion)
            cond4=(champion_data['Pos']==position.iloc[i])
            if j==0:
                lst1.extend(champion_data[cond1&cond2&cond3&cond4][champ_attr].iloc[0])
            else:
                lst2.extend(champion_data[cond1&cond2&cond3&cond4][champ_attr].iloc[0])
            i+=1
                

    

        j+=1
    after_data1.extend(lst1)
    after_data1.extend(lst2)
    after_data2.extend(lst2)
    after_data2.extend(lst1)
    data_body.append(after_data1)
    data_body.append(after_data2)
    
```

```python
export_data=pd.DataFrame(data=data_body, columns=columns)
```

```python
export_data.to_csv('./train_data/train_data.csv')
```
