```python
match_data=pd.read_csv("{0}match.csv".format(file_directory))
champion_data=pd.read_csv("{0}champion.csv".format(file_directory))
player_data=pd.read_csv("{0}player.csv".format(file_directory))
```

```python
match_data_index=set(match_data.index)
match_data.tail()
```

```python
lane=['Top','Jungle','Middle','ADC','Support']
player_attr=['W%','KDA','KP','GD10','XPD10','CSD10','DPM']
champ_attr=['P+B%','W%']
```

```python
columns=['side','result','P_W%','P_KDA','P_KP','P_GD10','P_XPD10','P_CSD10','P_DPM','C_P+B%','C_W%']
```

```python
data_body=[]
```

```python
for elem in match_data_index:
    
    temp1=[]
    temp2=[]
    lst1=[]
    lst2=[]
    after_data1=[]
    after_data2=[]
    temp_data=match_data.loc[elem][match_data.columns.difference(['league'])]
    datas=[temp_data[0:5],temp_data[5:]]
    
 
    for i,data in enumerate(datas):
            
             
        result=data.iloc[0]['result']
        side=data.iloc[0]['side']
        team=data.iloc[0]['team']
        year=data.iloc[0]['year']-2000
        season=data.iloc[0]['split']
        players=data['player']
        champions=data['champion']
        position=data['position']
       
      
        if i==0:
            after_data1=[result,side]
        else:
            after_data2=[result,side]
            
        
        cond1=(player_data['Year']==year)
        cond2=(player_data['Season']==season)
        cond5=(player_data['Team']==team)
        
        
 
          

        player_lst1=[]
        player_lst2=[]
        for j,player in enumerate(players):
            cond3=(player_data['Player']==player)
            cond4=(player_data['Pos']==position.iloc[j])
            if i==0:
                player_lst1.append(np.array(player_data[cond1&cond2&cond3&cond4&cond5][player_attr].iloc[0]))
            else:
                player_lst2.append(np.array(player_data[cond1&cond2&cond3&cond4&cond5][player_attr].iloc[0]))   
        
        
        
        if i==0:
            temp1=np.array(player_lst1)
            temp1=list(temp1.mean(axis=0))
        else:
            temp2=np.array(player_lst2)
            temp2=list(temp2.mean(axis=0))

    

        
        champ_lst1=[]
        champ_lst2=[]
        cond1=(champion_data['Year']==year)
        cond2=(champion_data['Season']==season)
        
        for j,champion in enumerate(champions):
            cond3=(champion_data['Champion']==champion)
            cond4=(champion_data['Pos']==position.iloc[j])
            if i==0:
                champ_lst1.append(champion_data[cond1&cond2&cond3&cond4][champ_attr].iloc[0])
            else:
                champ_lst2.append(champion_data[cond1&cond2&cond3&cond4][champ_attr].iloc[0])
                
        if i==0:
            temp1_1=np.array(champ_lst1)
            temp1.extend(list(temp1_1.mean(axis=0)))
        else:
            temp2_1=np.array(champ_lst2)
            temp2.extend(list(temp2_1.mean(axis=0)))
            
            
    temp1=np.array(temp1)
    temp2=np.array(temp2)
    after_data1.extend(list(temp1-temp2))
    after_data2.extend(list(temp2-temp1))
    data_body.append(after_data1)
    data_body.append(after_data2)

    
```

```python
export_data=pd.DataFrame(data=data_body, columns=columns)
```

```python
export_data.to_csv('./train_data/train_data1.csv')
```
